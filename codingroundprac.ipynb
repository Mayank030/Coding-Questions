```python
# Q8) Given 2 strings s and t, return if t is anagram of s and false otherwise.

# Example 

# s = "anagram"   t = "nagaram"


def is_anagram(s,t) :
    


    s_list = []
    t_list = []

    if len(s) != len(t):
        return False
    else: 

        for i in s:
            s_list.append(i)
            s_list.sort()
    
        for j in t:
            t_list.append(j)
            t_list.sort()

        if s_list == t_list: 
            return True 
    return False

        
        

print(is_anagram("anagram" , "nagaram"))
print(is_anagram("rock", "cork"))
print(is_anagram("Hair", "Body"))
```

```python
# Q10) Rearrange the array such that the first half is arranged in increasing order, and the second half is arranged in decreasing order.

# Input - 8,7,1,6,5,9

# Output - 1,5,6,9,8,7


def rearrange(arr):
    result = []
    n = len(arr)
    arr.sort()

    # for ascending
    for i in range(n//2):
        result.append(arr[i])
    
    # for descending
    for j in reversed(range(n//2, n)):
        result.append(arr[j])

    return result


print(rearrange([8,7,1,6,5,9]))
```

```python
# Q5) Print all Prime factors of given number.

# Example : 315 = 3 x 3 x 5 x 7

def prime_Factors(num):

    factors = []

    # The smallest prime number is 2, so we'll divide by 2 
    # Print the number of two's that divide num

    while num % 2 == 0 :
        factors.append(2)
        num = num / 2

    # We printed all the possible 2's
    # Now we'll check for rest of the prime numbers. (3,5,7...)
    # # N must be odd at this point, so we can skip even numbers
    for i in range(3,int(num**1/2)+1, 2):
        while num % i== 0:
            factors.append(i)
            num = num / i

    # This condition is to check if N is a prime number
    # greater than 2
    if num > 2:
        factors.append(num)

    return " ".join(map(str,factors))

print(prime_Factors(315))
print(prime_Factors(45))

```

```python
# Q6) GCD of two numbers

def gcd_of_two_numbers(a,b):

    # Everything divides 0
    if (a == 0):
        return b
    if (b == 0):
        return a

    # base case
    if (a == b):
        return a

    # a is greater
    if (a > b):
        return gcd_of_two_numbers(a-b, b)
    return gcd_of_two_numbers(a, b-a)

# Driver program to test above function
a = 98
b = 56
print(gcd_of_two_numbers(a,b))
print(gcd_of_two_numbers(4,16))
```

```python
# Q14) Check if a number is perfect number. 

# A number is said to be perfect if sum of all its factors excluding the number itself is equal to the number.

# Input:
# N = 6
# Output:
# 1 
# Explanation:
# Factors of 6 are 1, 2, 3 and 6.
# Excluding 6 their sum is 6 which is equal to N itself. So, it's a Perfect Number.


# Input:
# N = 10
# Output:
# 0
# Explanation:
# Factors of 10 are 1, 2, 5 and 10.
# Excluding 10 their sum is 8 which is not equal to N itself. So, it's not a Perfect Number.


def isPerfect(num): 
    factors = [1]
    for x in range(2,num):
        if num % x == 0:
            factors.append(x)

    if sum(factors) == num :
        return True 
    return False

print(isPerfect(6))
print(isPerfect(10))
print(isPerfect(28))

        
```

```python
# Q7) Find missing number in an array.

# You are given an array arr of size n - 1 that contains distinct integers in the range from 1 to n (inclusive). 
# This array represents a permutation of the integers from 1 to n with one element missing. 
# Your task is to identify and return the missing element.


# Input: arr[] = [8, 2, 4, 5, 3, 7, 1]
# Output: 6


def missing_ele(arr):
    n = len(arr) + 1
    summ_of_n_number = (1/2) * n * (n+1)
    summ_of_arr = sum(arr)
    missing_element = summ_of_n_number - summ_of_arr

    return int(missing_element)


print(missing_ele([8, 2, 4, 5, 3, 7, 1]))
print(missing_ele([1, 2, 3, 5]))
```

```python
# Q9) Move all zero to end of the array.

# Given an array arr[]. Push all the zeros of the given array to the right end of the array while maintaining the order of non-zero elements.
# Do the mentioned change in the array in place.

# Input: arr[] = [1, 2, 0, 4, 3, 0, 5, 0]
# Output: [1, 2, 4, 3, 5, 0, 0, 0]
# Explanation: There are three 0s that are moved to the end.


def moveZeroesToEnd(arr):
    
    zeroes = []
    non_zeroes = []

    
    for x in arr:
        if x > 0 :
            non_zeroes.append(x)
        else:
            zeroes.append(x)

    non_zeroes.extend(zeroes)

    return non_zeroes

print(moveZeroesToEnd([1, 2, 0, 4, 3, 0, 5, 0]))
print(moveZeroesToEnd([10,20,30]))
print(moveZeroesToEnd([0,0]))
```

```python
# Q11) Maximum Subarray sum (Kadanes Algorithm)

# Given an integer array arr[]. You need to find and return the maximum sum possible from all the subarrays.

# Input: arr[] = [2, 3, -8, 7, -1, 2, 3]
# Output: 11
# Explanation: The subarray {7, -1, 2, 3} has the largest sum 11.


# Input: arr[] = [-2, -4]
# Output: -2

# Input: arr[] = [5, 4, 1, 7, 8]
# Output: 25


def kadanesAlgo(arr):
    start = 1 

    max_curr = max_global = arr[0] 

    for i in range(start,len(arr)): 
        max_curr = max(arr[i], max_curr + arr[i])
        if max_curr > max_global: 
            max_global = max_curr
        start +=1 
    return max_global



print(kadanesAlgo([1, 2, 3, -8, 7, -1, 2, 3]))
print(kadanesAlgo([-2,-4]))
print(kadanesAlgo([5,4,1,7,8]))
```

```python
# Q5) Convert Lower to upper case and upper to lower case in a string. 

# ASCII values  of alphabets: A  Z = 65 to 90, a  z = 97 to 122 

# Steps: 

# 1. Take one string of any length and calculate its length.
# 2. Scan string character by character and keep checking the index. 
# If a character in an index is in lower case, then subtract 32 to convert it into upper case, else add 32 to convert it in lowercase

def lower_upper(strn):

    upperalphabets="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    loweralphabets="abcdefghijklmnopqrstuvwxyz"
    opposite_case = ""

    for i in strn:
        if i in upperalphabets:
            opposite_case += loweralphabets[upperalphabets.index(i)]
        else:
            opposite_case += upperalphabets[loweralphabets.index(i)]

    return opposite_case


print(lower_upper("HeLlo"))
```

```python
# Q6) Longest common prefix.

# Given an array of strings arr[]. Return the longest common prefix among each and every strings present in the array. 
# If there's no prefix common in all the strings, return "".

# Input: arr[] = ["geeksforgeeks", "geeks", "geek", "geezer"]
# Output: "gee"
# Explanation: "gee" is the longest common prefix in all the given strings.


def longestCommonPrefix(arr):
        
        arr.sort()   # ['geek', 'geeks', 'geeksforgeeks', 'geezer']
        
        first = arr[0]   # geek
    
        last = arr[-1]    # geezer
 
        minlength = min(len(first), len(last))    #  = 4
        
        i = 0
        while i < minlength and first[i] == last[i]:
            i += 1
        return first[:i]


print(longestCommonPrefix(["geeksforgeeks", "geeks", "geek", "geezer"]))
```
